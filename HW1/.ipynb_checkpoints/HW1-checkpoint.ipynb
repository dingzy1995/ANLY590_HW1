```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn import linear_model
```

```python
df = pd.read_csv("Hitters.csv")
df.head()
```

```python
# name = df['Unnamed: 0']
xs = df.drop(['Unnamed: 0', 'League', 'Division', 'NewLeague', 'Salary'], axis = 1)
xs_test = xs.tail(50)
xs = xs[:272]
xs.fillna(method = 'pad')
y = df['Salary']
y_test = y.tail(50)
y_test = y_test.fillna(method = 'bfill')
y = y[:272]
y = y.fillna(method = 'bfill')
y.values.reshape(-1, 1)

lasso = Lasso(alpha=12)
y_lasso = lasso.fit(xs, y).predict(xs_test)

print(str(lasso.coef_))
plt.plot(lasso.coef_, label='Lasso coefficients')
plt.legend(loc='best')
plt.show()
```

The 3 final three predictors that remain in the model are Hits, HmRun and Walks

```python
lassocv = LassoCV()
lassocv.fit(xs, y)
alpha = lassocv.alpha_
print(alpha)

lasso = Lasso(alpha=alpha)
y_lasso = lasso.fit(xs, y).predict(xs_test)
#number of nonzero coefficients
n = np.sum(lasso.coef_ != 0)
print(str(n))
plt.plot(lasso.coef_, label='LassoCV coefficients')
plt.legend(loc='best')
plt.show()
```

By cross validation, the optimal value of the regularization penality is 445.147,9 predictors are left in
the model

```python
ridg = linear_model.Ridge (alpha = 15)
y_ridg = ridg.fit(xs, y).predict(xs_test)

print(str(ridg.coef_))

plt.plot(ridg.coef_, label='ridge coefficients')
plt.legend(loc='best')
plt.show()
```

The final 3 predictors of this model are Hits, HmRun and Years.

```python
ridgecv = linear_model.RidgeCV()
ridgecv.fit(xs, y)
alpha = ridgecv.alpha_
print(alpha)

ridg = linear_model.Ridge (alpha = alpha)
y_ridg = ridg.fit(xs, y).predict(xs_test)
n = np.sum(lasso.coef_ != 0)
print(str(n))
plt.plot(ridg.coef_, label='ridge coefficients')
plt.legend(loc='best')
plt.show()
```
